### Linear Regression
- Fit a line to a data set of observations
- Use this line to predict unobserved values
- You can use it to predict points in the future, the past, whatever. In fact time usually has nothing to do with it.

- Linear regression usually use "least squares"
- minimizes the squared-error between each point and the line
- the slope-intercept equation of a line y=mx+b
- the slope is the correlation between the two variables times the standard deviation in Y, all devided by the standard deviation in X.
- The intercept is the mean of Y minus the slope times the mean of X
- Least squares minimizes the sum of squared errors
- This is the same as maximizing the likelihood of the observed data if you start thinking of the problem in terms of probabilities and probability distribution functions
- this is sometimes called "maximum likelihood estimation"

### There are other ways:
- Gradient Descent is an alternative method to least squares
- Basically iterates to find the line that best follows the contours defined by the data
- Can make sense when dealing with 3D data
- Easy to try in Python and just compare the results to least squares, but usually least squares is a perfectly good choice.


### Measuring error with r-squared
- How do we measure how well our line fits our data?
- R-squared (aka coefficient of determination) measured:
**The fraction of the total variation in Y that is captured by the model**

Computing r-squared: $1.0-sum\_of\_squared\_errors \div sum\_of\_squared\_variation\_from\_mean$

### Interpreting r-squared
- Ranges from 0 to 1
- 0 is bad (none of the variance is captured) - poor fit, 1 is good (all of the variance is captures) - good fit.

### Linear Regression Samples
Generation of data which represents linear relationship between page speed and amount purchased

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
page_speeds = np.random.normal(3.0, .5, 1000)
purchase_amount = 100 - (page_speeds + np.random.normal(0, 0.1, 1000)) *3
plt.scatter(page_speeds, purchase_amount)
```

We have only two features, we can keep it simple and use scipy.state.linregress

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(page_speeds, purchase_amount)
```

```python
# R-squared value shows a really good fit:
r_value ** 2
```

```python
# Use slope and intercept we got from the regression to plot predicted values vs. observed:
def predict(x):
    return slope * x + intercept

fit_line = predict(page_speeds)
plt.scatter(page_speeds, purchase_amount)
plt.plot(page_speeds, fit_line, c='r')
plt.show()
```
