```python
# Correlation between believers and HDI level:
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

human_dev_indexes = numpy.array([0.829, 0.759, 0.845, 0.785, 0.853, 0.822, 0.834, 0.872, 0.899, 0.812, 0.830, 0.869, 0.881, 0.834, 0.917, 0.911, 0.881, 0.818, 0.881, 0.879, 0.777, 0.895, 0.892, 0.810, 0.874, 0.884, 0.915, 0.944, 0.900, 0.898, 0.861, 0.840])
percents_of_believers = numpy.array([95, 90, 90, 90, 81, 91, 80, 74, 73, 67, 61, 59, 54, 49, 48, 47, 44, 44, 43, 41, 40, 38, 38, 37, 37, 34, 34, 32, 31, 23, 19, 16])
percents_of_atheists = numpy.array([100-i for i in percents_of_believers])
correlation_coeff = numpy.corrcoef(percents_of_atheists, human_dev_indexes)[0, 1]

plt.xlabel('percents of atheists', fontsize=18)
plt.ylabel('human development indexes', fontsize=18)
plt.plot(percents_of_atheists, human_dev_indexes, 'ro')
plt.suptitle("Correlation parameter = %s"%correlation_coeff, fontsize=20)
# plt.axis([0, 6, 0, 20])
interp_func = interp1d(percents_of_atheists, human_dev_indexes, kind='cubic')
# interp_func = interp1d(percents_of_atheists, human_dev_indexes, kind='linear')

xnew = numpy.arange(percents_of_atheists.min(), percents_of_atheists.max(), 0.01)
ynew = interp_func(xnew)

# plt.plot(percents_of_atheists, interp_func(percents_of_atheists), color="red")

# plt.plot(xnew, ynew, color="red")

coefficients = numpy.polyfit(percents_of_atheists, human_dev_indexes, 1)
polynomal = numpy.poly1d(coefficients)
ys = polynomal(percents_of_atheists)
plt.plot(percents_of_atheists, ys)

plt.show()
```
